## Part I – The Node class

### Task 1 – Define the class 

```python
class Node:
    def __init__(self, name ,neighbors=None):
        self.name = name
        self.neighbors  = {} # neighbors : edge_value
    
    def __str__(self):
        return "For {} neighbors are -- > {} ".format(self.name,self.neighbors)
    
    def __len__(self):
        return len(self.neighbors)
    
    def __eq__(self, other):
        return self.name == other.name
    
    def __ne__(self, other):
        return self.name != other.name
   
    def is_neighbor(self, name):
        if name in self.neighbors:
            return True
        else:
            return False
    
    def add_neighbor(self, name, weight=1):
        if name in self.neighbors.keys():
            return print('Neighbor name : {} is already exits'.format(name)) 
        if name==self.name:
            return print('Neighbor name : {} is the self.node'.format(name))
        else:
            self.neighbors[name]=weight
    
    def remove_neighbor(self, name):
        try:
            self.neighbors.pop(name)
            return 
        except KeyError:
            return print('{} is not neighbor of {}'.format(name,self.name))

        
    def get_weight(self, name):
        if self.is_neighbor(name):
            return self.neighbors[name]
         
    
    def is_isolated(self):
        return any(self.neighbors.keys())!=True
        
```

## Part II – The Graph class

### Task 1 – Define the class

```python
# part 2 :
class Graph:
    def __init__(self, name, nodes_list=[]):
        self.name = name
        self.nodes = {} # node name: Node object
        for node in nodes_list:
            self.nodes[node.name]=node
    
    def __str__(self):
        s_graph = self.name + ":\n\t"
        s_nodes = '\n\t'.join([str(node) for node in self.nodes.values()])
        return s_graph + s_nodes
    
    def __len__(self):
        return len(self.nodes)
    
    def __contains__(self, key):
        pass
    
    def __getitem__(self, name) :
        try:
            return self.nodes[name] in self.nodes
        except KeyError:
            print('{} is not a node in the graph'.format(name))    
    
    
    def is_edge(self, frm_name, to_name):
        if ((frm_name in self.nodes) and (to_name in self.nodes)):
            return self.nodes[frm_name].is_neighbor(to_name)
        else:
            return print("{} or {} are note a nodes in the graph".format(frm_name, to_name))
    
    
    def remove_node(self, name):
        try:
            self.nodes.pop(name)
        except KeyError:
            print('{} is not a node of grpah :{}'.format(name,self.name))
    
   
    def add_node(self, node):
        if node.name in self.nodes:
            for name_neighbor, weight in node.neighbors.items():
                if not self.nodes[node.name].is_neighbor(name_neighbor):
                    self.nodes[node.name].add_neighbor(name_neighbor,weight)
        else:
            self.nodes[node.name]=node
    
    
    def __add__(self, other):
        join_name = "{} + {}".format(self.name, other.name)
        join_grpah = Graph(join_name, list(self.nodes.values()))
        for node in other.nodes.values():
            join_grpah.add_node(node)
        return join_grpah
    
    def add_edge(self, frm_name, to_name, weight=1):
        if (frm_name not in self.nodes or to_name not in self.nodes):
            print("One or more of the nodes [{}, {}] are not in the graph".format(frm_name,to_name))
            return
        if self.nodes[frm_name].is_neighbor(to_name):
            print("{} is already a neighbor of {}".format(frm_name,to_name))
            return
        if self.nodes[frm_name]==self.nodes[to_name]:
            print("The nodes are identical")
            return
        return self.nodes[frm_name].add_neighbor(to_name,weight)
    
    def remove_edge(self, frm_name, to_name) :
        if (frm_name not in self.nodes ):
            print("{} is not a node in the graph".format(frm_name))
            return
    
        if self.nodes[frm_name].is_neighbor(to_name)==False:
            print("{} is not a neighbor of {}".format(to_name,frm_name))
            return
        
        return self.nodes[frm_name].remove_neighbor(to_name)
        
    
    def get_edge_weight(self, frm_name, to_name):
        if (frm_name not in self.nodes or to_name not in self.nodes):
            print("One or more of the nodes [{}, {}] are not in the graph".format(frm_name,to_name))
            return 0
        if self.nodes[frm_name].is_neighbor(to_name)==False:
            print("{} is not a neighbor of {} (get_edge_weight method)".format(to_name,frm_name))
            return 0
        if self.nodes[frm_name].is_neighbor(to_name):
            return self.nodes[frm_name].get_weight(to_name)
    
    def get_path_weight(self, path):
        if any(path)==False:
            print("Path is empty")
            return
        if self.is_reachable(path[0],path[-1])==False:
            print("No path between {} to {}".format(path[0],path[-1]))
        #path=[8,7,6]
        queue=path.copy()
        tot_weight=0
        while len(queue)>1:
            frm_name=queue[0]
            to_name=queue[1]
            tot_weight=tot_weight+self.get_edge_weight(frm_name,to_name)
            queue.pop(0)
        return tot_weight
    
                
    def is_reachable(self, start, end):
        stack = [start]
        path =  []
        
        while stack:
            node=stack.pop()
            for neighbor in self.nodes[node].neighbors.keys():
                if neighbor in path:
                    continue
                else:
                    path.append(neighbor)
                    stack.append(neighbor)
        if end in path:
#             print(path)
            return True
        else:
            return False
        
    def reachable_nodes(self, start):
        stack = [start]
        path =  []
        
        while stack:
            node=stack.pop()
            for neighbor in self.nodes[node].neighbors.keys():
                if neighbor in path:
                    continue
                else:
                    path.append(neighbor)
                    stack.append(neighbor)
        if start in path:
            path.remove(start)
        return path
    
    def find_shortest_path(self, start, end):
        #print('find_shortest_path from {} to {}'.format(start, end))
        stack = []
        stack.append([start])
        path_weight=0
        
        while stack:
            list_test=stack.pop()
            node=list_test[-1]
            ## in case node don't have neighbors and we reach to the end of path
            if node==end:
                copy_list_test = list_test.copy()
                if ((self.get_path_weight(copy_list_test)<path_weight) or (path_weight==0)):
                    path_weight = self.get_path_weight(copy_list_test)
                    minimal_path_size = len(copy_list_test)
                    #print('found path {} with score {} and minimal path size: {}'.format(path_weight, copy_list_test, minimal_path_size))
            elif (path_weight>0 and self.get_path_weight(list_test)>path_weight):
                continue
            else:
                for neighbor in self.nodes[node].neighbors.keys():
                    if neighbor in list_test:
                        continue
                    copy_list_test=list_test.copy()
                    copy_list_test.append(neighbor)
                    stack.append(copy_list_test)
        return path_weight    
    
```

### Part III – Non-directional graph

### Task 1 – define the class
* Implement the NonDirectionalGraph class as a sub-class of Graph. The main property of the non-directional graph is that its edges come in pairs, so if an edge is added or remove, then the class must make sure that the same applies to its counterpart. There are no additional methods to implement, but make sure to overwrite the relevant methods.


```python
class NonDirectionalGraph(Graph):
    
    def add_edge(self, frm_name, to_name, weight=1):
        Graph.add_edge(self,frm_name, to_name, weight)
        
        self.nodes[to_name].add_neighbor(frm_name,weight)
        
    def remove_edge(self, frm_name, to_name, weight=1):
        Graph.add_edge(self,frm_name, to_name)
        
        self.nodes[to_name].remove_neighbor(frm_name)
        
    def add_node(self, node):
        if node.name in self.nodes:
            for name_neighbor, weight in node.neighbors.items():
                if not self.nodes.is_edge(node.name, name_neighbor):
                    self.nodes[node.name].add_neighbor(name_neighbor,weight)
                    self.nodes[name_neighbor].add_neighbor(node.name,weight)
        else:
            self.nodes[node.name]=node
    
    
    
```

```python
network = NonDirectionalGraph('social',[])
list_help=[]
n_edges=[0]
with open("social.txt", 'r') as f:
    for line in f:
        line_as_list = line.split(' ')
        list_help.append((line_as_list[0],line_as_list[2],line_as_list[3]))
    for item in list_help:
        if item[2]=='became':
            network.add_node(Node(item[0]))
            network.add_node(Node(item[1]))
            network.add_edge(item[0],item[1])
            n_edges.append((n_edges[-1]+1))
        else:
            network.remove_edge(item[0],item[1])
            n_edges.append((n_edges[-1]-1))
print(max(n_edges))
            
```
